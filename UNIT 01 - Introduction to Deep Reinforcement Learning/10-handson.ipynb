- `truncated`: Bu yeni sÃ¼rÃ¼mle birlikte tanÄ±tÄ±ldÄ±, bir zaman sÄ±nÄ±rÄ±nÄ± veya Ã¶rneÄŸin bir ajanÄ±n ortam sÄ±nÄ±rlarÄ±nÄ±n dÄ±ÅŸÄ±na Ã§Ä±kÄ±p Ã§Ä±kmadÄ±ÄŸÄ±nÄ± gÃ¶sterir.

Hadi bir Ã¶rneÄŸe bakalÄ±m! Kodu okuduÄŸunuzdan emin olun

```python
import gymnasium as gym

# First, we create our environment called LunarLander-v2
env = gym.make("LunarLander-v2")

# Then we reset this environment
observation, info = env.reset()

for _ in range(20):
    # Take a random action
    action = env.action_space.sample()
    print("Action taken:", action)
    
    # Do this action in the environment and get
    # next_state, reward, terminated, truncated and info
    observation, reward, terminated, truncated, info = env.step(action)
    
    # If the game is terminated (in our case we land, crashed) or truncated (timeout)
    if terminated or truncated:
        # Reset the environment
        print("Environment is reset")
        observation, info = env.reset()

env.close()
```

## Modeli OluÅŸturun ğŸ¤–

- Ã‡evremizi inceledik ve sorunu anladÄ±k: **Lunar Lander'Ä± sol, saÄŸ ve ana yÃ¶nlendirme motorunu kontrol ederek Ä°niÅŸ Pistine doÄŸru ÅŸekilde indirebilmek**. Åimdi bu Problemi Ã§Ã¶zmek iÃ§in kullanacaÄŸÄ±mÄ±z algoritmayÄ± oluÅŸturalÄ±m ğŸš€.

- Bunu yapmak iÃ§in, ilk Derin RL kÃ¼tÃ¼phanemiz olan [Stable Baselines3 (SB3)] (https://stable-baselines3.readthedocs.io/en/master/) kullanacaÄŸÄ±z.

- SB3, PyTorch'ta takviye Ã¶ÄŸrenme algoritmalarÄ±nÄ±n **gÃ¼venilir uygulamalarÄ±ndan oluÅŸan bir settir**.

---

ğŸ’¡ Yeni bir kÃ¼tÃ¼phane kullanÄ±rken iyi bir alÄ±ÅŸkanlÄ±k, Ã¶nce belgelere dalmaktÄ±r: https://stable-baselines3.readthedocs.io/en/master/ ve ardÄ±ndan bazÄ± Ã¶ÄŸreticileri deneyin.

![image3](https://huggingface.co/datasets/huggingface-deep-rl-course/course-images/resolve/main/en/unit1/sb3.png)

Bu sorunu Ã§Ã¶zmek iÃ§in SB3 **PPO** kullanacaÄŸÄ±z. [PPO (diÄŸer adÄ±yla Proximal Policy Optimization), bu kurs sÄ±rasÄ±nda Ã§alÄ±ÅŸacaÄŸÄ±nÄ±z SOTA ("state of the art") Deep Reinforcement Learning algoritmalarÄ±ndan biridir] (https://stable-baselines3.readthedocs.io/en/master/modules/ppo.html#example%5D).

PPO ÅŸunlarÄ±n bir kombinasyonudur:
- *DeÄŸer tabanlÄ± takviye Ã¶ÄŸrenme yÃ¶ntemi*: bize bir durum ve eylem verildiÄŸinde yapÄ±lacak **en deÄŸerli eylemi sÃ¶yleyecek bir eylem-deÄŸer fonksiyonu Ã¶ÄŸrenmek**.
- *Politika tabanlÄ± takviye Ã¶ÄŸrenme yÃ¶ntemi*: **bize eylemler Ã¼zerinde bir olasÄ±lÄ±k daÄŸÄ±lÄ±mÄ± verecek** bir politika Ã¶ÄŸrenmek.

Stable-Baselines3'Ã¼n kurulumu kolaydÄ±r:

1ï¸âƒ£ Siz **ortamÄ±nÄ±zÄ± oluÅŸturun** (bizim durumumuzda bu yukarÄ±da yapÄ±ldÄ±)

2ï¸âƒ£ Kullanmak istediÄŸiniz **modeli tanÄ±mlar ve bu modeli Ã¶rneklendirirsiniz** `model = PPO("MlpPolicy")`

3ï¸âƒ£ AjanÄ± `model.learn` ile **eÄŸitirsiniz** ve eÄŸitim zaman adÄ±mlarÄ±nÄ±n sayÄ±sÄ±nÄ± tanÄ±mlarsÄ±nÄ±z

```python
# Create environment
env = gym.make('LunarLander-v2')

# Instantiate the agent
model = PPO('MlpPolicy', env, verbose=1)

# Train the agent
model.learn(total_timesteps=int(2e5))
```

`verbose=1` parametresi ayrÄ±ntÄ± dÃ¼zeyini veya PPO aracÄ±sÄ±nÄ±n Ã§alÄ±ÅŸmasÄ± sÄ±rasÄ±nda Ã§Ä±ktÄ± olarak verdiÄŸi bilgi miktarÄ±nÄ± kontrol eder.

`verbose=0`: Ã‡Ä±ktÄ± yok. AracÄ±, eÄŸitim sÄ±rasÄ±nda herhangi bir bilgi yazdÄ±rmayacaktÄ±r. Bu, eÄŸitim sÃ¼recini sessizce yÃ¼rÃ¼tmek istiyorsanÄ±z kullanÄ±ÅŸlÄ±dÄ±r.

`verbose=1`: Bilgi seviyesi. Ajan, eÄŸitim sÄ±rasÄ±nda bÃ¶lÃ¼m numaralarÄ± ve Ã¶dÃ¼ller de dahil olmak Ã¼zere eÄŸitimin ilerleyiÅŸi gibi bazÄ± bilgileri yazdÄ±racaktÄ±r. Bu, kullanÄ±cÄ±yÄ± Ã§ok fazla detayla boÄŸmadan eÄŸitim sÃ¼recini anlamak iÃ§in yeterli bilgiye sahip olmak arasÄ±nda iyi bir denge saÄŸlar.

`verbose=2`: Hata ayÄ±klama seviyesi. AracÄ±, eÄŸitim sÄ±rasÄ±nda ayrÄ±ntÄ±lÄ± bilgiler yazdÄ±racaktÄ±r. Bu, Ã¶ÄŸrenme sÃ¼reci, hiperparametreler ve diÄŸer dahili durumlar hakkÄ±nda daha ayrÄ±ntÄ±lÄ± ayrÄ±ntÄ±lar iÃ§erebilir. Bu ayrÄ±ntÄ± dÃ¼zeyi, hata ayÄ±klama veya algoritmanÄ±n iÃ§ iÅŸleyiÅŸini anlamak iÃ§in kullanÄ±ÅŸlÄ±dÄ±r.

```python
model = PPO(
    policy="MlpPolicy",
    env=env,
    n_steps=1024,
    batch_size=128,
    n_epochs=5,
    gamma=0.999,
    gae_lambda=0.98,
    ent_coef=0.01,
    verbose=1,
)
```

## Train the PPO agent ğŸƒ

- AjanÄ±mÄ±zÄ± 1.000.000 zaman adÄ±mÄ± iÃ§in eÄŸitelim, Colab Ã¼zerinde GPU kullanmayÄ± unutmayÄ±n. YaklaÅŸÄ±k ~20 dakika sÃ¼recektir, ancak sadece denemek istiyorsanÄ±z daha az zaman adÄ±mÄ± kullanabilirsiniz.

```python
model.learn(total_timesteps=1_000_000)

model_name = "ppo-LunarLander-v2"
model.save(model_name)
```

## AjanÄ± deÄŸerlendirin ğŸ“ˆ

- OrtamÄ± bir [Monitor](https://stable-baselines3.readthedocs.io/en/master/common/monitor.html) iÃ§ine sarmayÄ± unutmayÄ±n.
- ArtÄ±k Lunar Lander ajanÄ±mÄ±z eÄŸitildiÄŸine gÃ¶re, **performansÄ±nÄ± kontrol etmemiz** gerekiyor.
- Stable-Baselines3 bunu yapmak iÃ§in bir yÃ¶ntem saÄŸlar: `evaluate_policy`.
- Bu kÄ±smÄ± doldurmak iÃ§in [belgeleri kontrol edin](https://stable-baselines3.readthedocs.io/en/master/guide/examples.html#basic-usage-training-saving-loading)
- Bir sonraki adÄ±mda, **bir liderlik tablosunda yarÄ±ÅŸmak iÃ§in temsilcinizi otomatik olarak nasÄ±l deÄŸerlendireceÄŸinizi ve paylaÅŸacaÄŸÄ±nÄ±zÄ± gÃ¶receÄŸiz, ancak ÅŸimdilik bunu kendimiz yapalÄ±m**


ğŸ’¡ Temsilcinizi deÄŸerlendirirken eÄŸitim ortamÄ±nÄ±zÄ± kullanmamalÄ±, bir deÄŸerlendirme ortamÄ± oluÅŸturmalÄ±sÄ±nÄ±z.

```python
eval_env = Monitor(gym.make("LunarLander-v2"))
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, deterministic=True)
print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```
