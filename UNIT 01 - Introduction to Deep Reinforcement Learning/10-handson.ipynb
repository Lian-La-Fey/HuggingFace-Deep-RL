## Understand Gymnasium and how it works 🤖

🏋 Ortamımızı içeren kütüphanenin adı Gymnasium. Gymnasium'u Deep Reinforcement Learning'de çok kullanacaksınız.

Gymnasium, Farama Vakfı tarafından sürdürülen Gym kütüphanesinin yeni sürümüdür.

Gymnasium kütüphanesi iki şey sağlar:

- RL ortamları oluşturmanızı sağlayan bir arayüz. 
- Bir ortam koleksiyonu (gym-control, atari, box2D...). 

Bir örneğe bakalım, ama önce RL döngüsünü hatırlayalım.

![image2](https://huggingface.co/datasets/huggingface-deep-rl-course/course-images/resolve/main/en/unit1/RL_process_game.jpg)

Her adımda:

- Ajanımız Ortamdan bir durum (S0) alır - oyunumuzun (Ortam) ilk karesini alırız.
- Bu duruma (S0) dayanarak, Ajan bir eylemde bulunur (A0) - Ajanımız sağa doğru hareket edecektir.
- Ortam yeni bir duruma (S1) geçer - yeni çerçeve.
- Ortam, Ajana bir ödül (R1) verir - ölmedik (Pozitif Ödül +1).

Gymnasium ile:

1️⃣ `gymnasium.make()` kullanarak ortamımızı oluşturuyoruz

2️⃣ observation = `env.reset()` ile ortamı ilk durumuna sıfırlıyoruz

Her adımda:

3️⃣ Modelimizi kullanarak bir eylem alın (örneğimizde rastgele bir eylem alıyoruz)

4️⃣ `env.step(action)` kullanarak, bu eylemi ortamda gerçekleştirir ve şunları elde ederiz

- `observation`: Yeni durum (st+1)
- `reward`: Eylemi gerçekleştirdikten sonra aldığımız ödül
- `terminated`: Bölümün sonlandırılıp sonlandırılmadığını belirtir (aracı terminal durumuna ulaşır)
- `truncated`: Bu yeni sürümle birlikte tanıtıldı, bir zaman sınırını veya örneğin bir ajanın ortam sınırlarının dışına çıkıp çıkmadığını gösterir.
- `info`: Ek bilgi sağlayan bir sözlük (ortama bağlıdır).

Bölüm sonlandırılırsa:

- observation = `env.reset()` ile ortamı ilk durumuna sıfırlıyoruz

Hadi bir örneğe bakalım! Kodu okuduğunuzdan emin olun

```python
import gymnasium as gym

# First, we create our environment called LunarLander-v2
env = gym.make("LunarLander-v2")

# Then we reset this environment
observation, info = env.reset()

for _ in range(20):
    # Take a random action
    action = env.action_space.sample()
    print("Action taken:", action)
    
    # Do this action in the environment and get
    # next_state, reward, terminated, truncated and info
    observation, reward, terminated, truncated, info = env.step(action)
    
    # If the game is terminated (in our case we land, crashed) or truncated (timeout)
    if terminated or truncated:
        # Reset the environment
        print("Environment is reset")
        observation, info = env.reset()

env.close()
```

Eylem uzayı (temsilcinin gerçekleştirebileceği olası eylemler kümesi), mevcut 4 eylem ile ayrıktır 🎮:

Ödül fonksiyonu (her zaman adımında bir ödül verecek fonksiyon) 💰:

Her adımdan sonra bir ödül verilir. Bir bölümün toplam ödülü **o bölümdeki tüm adımların ödüllerinin toplamıdır**.

Her adım için ödül

- İniş aracı iniş pistine yaklaştıkça/uzaklaştıkça artar/azalır.
- İniş aracı ne kadar yavaş/hızlı hareket ediyorsa o kadar artar/azalır.
- İniş aracı ne kadar eğilirse (yatay olmayan açı) o kadar azalır.
- Yere temas eden her bacak için 10 puan artar.
- Bir yan motorun ateşlendiği her kare için 0,03 puan azalır.
- Ana motorun ateşlendiği her kare için 0,3 puan azaltılır.

Bölüm, çarpma veya güvenli iniş için sırasıyla -100 veya +100 puanlık bir **ek ödül alır.**

## Modeli Oluşturun 🤖

- Çevremizi inceledik ve sorunu anladık: **Lunar Lander'ı sol, sağ ve ana yönlendirme motorunu kontrol ederek İniş Pistine doğru şekilde indirebilmek**. Şimdi bu Problemi çözmek için kullanacağımız algoritmayı oluşturalım 🚀.

- Bunu yapmak için, ilk Derin RL kütüphanemiz olan [Stable Baselines3 (SB3)] (https://stable-baselines3.readthedocs.io/en/master/) kullanacağız.

- SB3, PyTorch'ta takviye öğrenme algoritmalarının **güvenilir uygulamalarından oluşan bir settir**.

---

💡 Yeni bir kütüphane kullanırken iyi bir alışkanlık, önce belgelere dalmaktır: https://stable-baselines3.readthedocs.io/en/master/ ve ardından bazı öğreticileri deneyin.

----

![image3](https://huggingface.co/datasets/huggingface-deep-rl-course/course-images/resolve/main/en/unit1/sb3.png)

Bu sorunu çözmek için SB3 **PPO** kullanacağız. [PPO (diğer adıyla Proximal Policy Optimization), bu kurs sırasında çalışacağınız SOTA ("state of the art") Deep Reinforcement Learning algoritmalarından biridir] (https://stable-baselines3.readthedocs.io/en/master/modules/ppo.html#example%5D).

PPO şunların bir kombinasyonudur:
- *Değer tabanlı takviye öğrenme yöntemi*: bize bir durum ve eylem verildiğinde yapılacak **en değerli eylemi söyleyecek bir eylem-değer fonksiyonu öğrenmek**.
- *Politika tabanlı takviye öğrenme yöntemi*: **bize eylemler üzerinde bir olasılık dağılımı verecek** bir politika öğrenmek.

Stable-Baselines3'ün kurulumu kolaydır:

1️⃣ Siz **ortamınızı oluşturun** (bizim durumumuzda bu yukarıda yapıldı)

2️⃣ Kullanmak istediğiniz **modeli tanımlar ve bu modeli örneklendirirsiniz** `model = PPO("MlpPolicy")`

3️⃣ Ajanı `model.learn` ile **eğitirsiniz** ve eğitim zaman adımlarının sayısını tanımlarsınız

```python
# Create environment
env = gym.make('LunarLander-v2')

# Instantiate the agent
model = PPO('MlpPolicy', env, verbose=1)

# Train the agent
model.learn(total_timesteps=int(2e5))
```

`verbose=1` parametresi ayrıntı düzeyini veya PPO aracısının çalışması sırasında çıktı olarak verdiği bilgi miktarını kontrol eder.

`verbose=0`: Çıktı yok. Aracı, eğitim sırasında herhangi bir bilgi yazdırmayacaktır. Bu, eğitim sürecini sessizce yürütmek istiyorsanız kullanışlıdır.

`verbose=1`: Bilgi seviyesi. Ajan, eğitim sırasında bölüm numaraları ve ödüller de dahil olmak üzere eğitimin ilerleyişi gibi bazı bilgileri yazdıracaktır. Bu, kullanıcıyı çok fazla detayla boğmadan eğitim sürecini anlamak için yeterli bilgiye sahip olmak arasında iyi bir denge sağlar.

`verbose=2`: Hata ayıklama seviyesi. Aracı, eğitim sırasında ayrıntılı bilgiler yazdıracaktır. Bu, öğrenme süreci, hiperparametreler ve diğer dahili durumlar hakkında daha ayrıntılı ayrıntılar içerebilir. Bu ayrıntı düzeyi, hata ayıklama veya algoritmanın iç işleyişini anlamak için kullanışlıdır.

```python
model = PPO(
    policy="MlpPolicy",
    env=env,
    n_steps=1024,
    batch_size=128,
    n_epochs=5,
    gamma=0.999,
    gae_lambda=0.98,
    ent_coef=0.01,
    verbose=1,
)
```

## Train the PPO agent 🏃

- Ajanımızı 1.000.000 zaman adımı için eğitelim, Colab üzerinde GPU kullanmayı unutmayın. Yaklaşık ~20 dakika sürecektir, ancak sadece denemek istiyorsanız daha az zaman adımı kullanabilirsiniz.

```python
model.learn(total_timesteps=1_000_000)

model_name = "ppo-LunarLander-v2"
model.save(model_name)
```

## Ajanı değerlendirin 📈

- Ortamı bir [Monitor](https://stable-baselines3.readthedocs.io/en/master/common/monitor.html) içine sarmayı unutmayın.
- Artık Lunar Lander ajanımız eğitildiğine göre, **performansını kontrol etmemiz** gerekiyor.
- Stable-Baselines3 bunu yapmak için bir yöntem sağlar: `evaluate_policy`.
- Bu kısmı doldurmak için [belgeleri kontrol edin](https://stable-baselines3.readthedocs.io/en/master/guide/examples.html#basic-usage-training-saving-loading)
- Bir sonraki adımda, **bir liderlik tablosunda yarışmak için temsilcinizi otomatik olarak nasıl değerlendireceğinizi ve paylaşacağınızı göreceğiz, ancak şimdilik bunu kendimiz yapalım**


💡 Temsilcinizi değerlendirirken eğitim ortamınızı kullanmamalı, bir değerlendirme ortamı oluşturmalısınız.

```python
eval_env = Monitor(gym.make("LunarLander-v2"))
mean_reward, std_reward = evaluate_policy(model, eval_env, n_eval_episodes=10, deterministic=True)
print(f"mean_reward={mean_reward:.2f} +/- {std_reward}")
```
