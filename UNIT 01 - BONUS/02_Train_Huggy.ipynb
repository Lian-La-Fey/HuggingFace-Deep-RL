## Let’s recap how this environment works

### The State Space: what Huggy perceives.

Huggy çevresini "görmüyor". Bunun yerine, ona çevre hakkında bilgi sağlarız:

- Hedefin (çubuğun) konumu
- Kendisi ile hedef arasındaki göreceli konum
- Bacaklarının yönü

Tüm bu bilgiler ışığında, Huggy hedefine ulaşmak için hangi eylemi gerçekleştireceğine karar verebilir.

![](https://huggingface.co/datasets/huggingface-deep-rl-course/course-images/resolve/main/en/notebooks/unit-bonus1/huggy.jpg)

### The Action Space: what moves Huggy can do

![](https://huggingface.co/datasets/huggingface-deep-rl-course/course-images/resolve/main/en/notebooks/unit-bonus1/huggy-action.jpg)

Eklem motorları Huggy'nin bacaklarını hareket ettirir. Bu da hedefe ulaşmak için Huggy'nin hareket edebilmesi için her bir bacağının eklem motorlarını doğru şekilde döndürmeyi öğrenmesi gerektiği anlamına geliyor.

### The Reward Function

Ödül fonksiyonu, Huggy'nin hedefini gerçekleştirmesi için tasarlanmıştır: sopayı getirmesi.

Takviyeli Öğrenmenin temellerinden birinin ödül hipotezi olduğunu unutmayın: bir hedef, beklenen kümülatif ödülün maksimizasyonu olarak tanımlanabilir.

Burada amacımız Huggy'nin sopaya doğru gitmesi ancak çok fazla dönmemesidir. Dolayısıyla, ödül fonksiyonumuz bu hedefi tercüme etmelidir.

Ödül fonksiyonumuz:

![](https://huggingface.co/datasets/huggingface-deep-rl-course/course-images/resolve/main/en/notebooks/unit-bonus1/reward.jpg)

- Yönlendirme bonusu: hedefe yaklaştığı için onu ödüllendiriyoruz.
- Zaman cezası: çubuğa mümkün olduğunca hızlı ulaşmaya zorlamak için her eylemde verilen sabit bir zaman cezası.
- Dönme cezası: çok fazla dönerse ve çok hızlı dönerse Huggy'yi cezalandırıyoruz.
- Hedefe ulaşma ödülü: Huggy'yi hedefe ulaştığı için ödüllendiriyoruz.



### Check the Huggy config file

- ML-Agents'da, config.yaml dosyalarında **eğitim hiperparametrelerini tanımlarsınız.**

- Bu not defterinin kapsamı için hiperparametreleri değiştirmeyeceğiz, ancak bir deney olarak denemek isterseniz, Unity çok [her birini burada açıklayan iyi belgeler] sağlar (https://github.com/Unity-Technologies/ml-agents/blob/main/docs/Training-Configuration-File.md).

- Huggy için bir yapılandırma dosyası oluşturmamız gerekiyor.

- Content/ml-agents/config/ppo` adresine gidin

- Huggy.yaml` adında yeni bir dosya oluşturun

```python
os.chdir('/content/ml-agents/config/ppo')
os.getcwd()
```

```python
import yaml

data = {
    'behaviors': {
        'Huggy': {
            'trainer_type': 'ppo',
            'hyperparameters': {
                'batch_size': 2048,
                'buffer_size': 20480,
                'learning_rate': 0.0003,
                'beta': 0.005,
                'epsilon': 0.2,
                'lambd': 0.95,
                'num_epoch': 3,
                'learning_rate_schedule': 'linear'
            },
            'network_settings': {
                'normalize': True,
                'hidden_units': 512,
                'num_layers': 3,
                'vis_encode_type': 'simple'
            },
            'reward_signals': {
                'extrinsic': {
                    'gamma': 0.995,
                    'strength': 1.0
                }
            },
            'checkpoint_interval': 200000,
            'keep_checkpoints': 15,
            'max_steps': 2e6,
            'time_horizon': 1000,
            'summary_freq': 50000
        }
    }
}

with open('Huggy.yaml', 'w') as file:
    yaml.dump(data, file, default_flow_style=False)
```

## Train our agent

Ajanımızı eğitmek için tek yapmamız gereken **mlagents-learn'i başlatmak ve ortamı içeren yürütülebilir dosyayı seçmektir.**

![image3](https://huggingface.co/datasets/huggingface-deep-rl-course/course-images/resolve/main/en/notebooks/unit-bonus1/mllearn.png)

ML Agent'lar ile bir eğitim komut dosyası çalıştırıyoruz. Dört parametre tanımlıyoruz:

1. `mlagents-learn <config>`: hiperparametre yapılandırma dosyasının bulunduğu yol.
2. `--env`: ortam çalıştırılabilir dosyasının bulunduğu yer.
3. `--run-id`: eğitim çalıştırma kimliğinize vermek istediğiniz isim.
4. `no-graphics`: eğitim sırasında görselleştirmeyi başlatmamak için.

Modeli eğitin ve kesinti durumunda eğitime devam etmek için `--resume` bayrağını kullanın.

> `--resume` kullandığınızda ilk seferde başarısız olacaktır, hatayı atlamak için bloğu tekrar çalıştırmayı deneyin.
